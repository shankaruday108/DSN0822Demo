```python
# Now connection to the driver which is available in this folder of my laptop
driver = webdriver.Chrome(r"D:\Uday Shankar's Data\Softwares\chromedriver_win32\chromedriver.exe")
```

```python
# Now opening the naukri.com website in a new window
driver.get("https://www.imdb.com/search/title/?genres=action&sort=user%20rating.desc&title%20type=feature&nu%20m%20votes=25000,&pf%20rd%20m=A2FGELUUNOQJNL&pf%20rd%20p=f11158cc-b50b-4c4d-b0a2-%2040b32863395b&pf%20rd%20r=XZ8X52H1R40B7KG5SNZ9&pf%20rd%20s=right-%206&pf%20rd%20t=15506&pf_rd_i=top&ref_=chttp_gnr%C2%A01")
driver.fullscreen_window()
time.sleep(3)
```

```python
# getting All products
all_products_ud  = driver.find_elements(By.XPATH, "//div[@class='lister-item-content']")
time.sleep(3)
len(all_products_ud)

movie_serial_numbers = []
movie_metascores = []
movie_ratings = []
movie_names = []
movie_years = []
movie_votes = []
movie_gross = []
movie_durations = []
movie_certificates = []
movie_genres = []

movie_directors = []
movie_director_1 = []
movie_director_2 = []
movie_director_3 = []

movie_star_1 = []
movie_star_2 = []
movie_star_3 = []
movie_star_4 = []

for allUd in all_products_ud[0:len(all_products_ud)]:
    
    try:
        
        # getting movies Serial Numbers
        if len(allUd.find_elements(By.XPATH, ".//span[@class='lister-item-index unbold text-primary']")) > 0:
            all_serial_numbers  = allUd.find_elements(By.XPATH, ".//span[@class='lister-item-index unbold text-primary']")

            for i in all_serial_numbers:
                
                movie_serial_numbers.append(i.text.strip('.'))
                
        else:
            movie_serial_numbers.append("--")
            
    except:
        pass               
            
    try:    
        
        # getting movies All metascores
        if len(allUd.find_elements(By.XPATH, ".//div[@class='inline-block ratings-metascore']/span")) > 0:
            all_metascores  = allUd.find_elements(By.XPATH, ".//div[@class='inline-block ratings-metascore']/span")

            for i in all_metascores:
                movie_metascores.append(i.text)
                
        else:
            movie_metascores.append("--")       
                
    except:
        pass
    
    
    try:
        
        # getting movies All ratings
        if len(allUd.find_elements(By.XPATH, ".//div[@class='inline-block ratings-imdb-rating']/strong")) > 0:
            all_ratings  = allUd.find_elements(By.XPATH, ".//div[@class='inline-block ratings-imdb-rating']/strong")

            for i in all_ratings:
                movie_ratings.append(i.text)
        else:
            movie_ratings.append("--") 
            
    except:
        pass
    
    try:
        
        # getting movies name
        if len(allUd.find_elements(By.XPATH, ".//h3/a")) > 0:
            brand_names  = allUd.find_elements(By.XPATH, ".//h3/a")
            
            for i in brand_names:
                movie_names.append(i.text.replace(',', ''))
                
        else:
            movie_names.append("--")       
        
    except:
        pass
    
    try:
        
        # getting movies years values
        if len(allUd.find_elements(By.XPATH, ".//span[@class='lister-item-year text-muted unbold']")) > 0:
            all_years  = allUd.find_elements(By.XPATH, ".//span[@class='lister-item-year text-muted unbold']")

            for i in all_years:
                movie_years.append(i.text)
        else:
            movie_years.append("--")             
        
    except:
        pass
    
    try:
        
        # getting movies All Votes
        if len(allUd.find_elements(By.XPATH, ".//p[@class='sort-num_votes-visible']//span[2]")) > 0:
            all_votes  = allUd.find_elements(By.XPATH, ".//p[@class='sort-num_votes-visible']//span[2]")

            for i in all_votes:
                movie_votes.append(i.text.replace(',', ''))
                    
        else:
            movie_votes.append("--")            

    except:
        pass
    
    try:
        
        # getting movies All Gross
        if len(allUd.find_elements(By.XPATH, ".//p[@class='sort-num_votes-visible']//span[5]")) > 0:
            all_gross  = allUd.find_elements(By.XPATH, ".//p[@class='sort-num_votes-visible']//span[5]")

            for i in all_gross:
                movie_gross.append(i.text.replace(',', ''))
                    
        else:
            movie_gross.append("--")            

    except:
        pass
    
    
    try:
        
        # getting movies All Directors name
        
        all_directors  = allUd.find_elements(By.XPATH, ".//p[3]")

        for i in all_directors:
            if i.text:
                director = i.text.split(' | ')[0]

                if re.search("Director:", director):   
                    movie_directors.append(director)
                    movie_director_1.append(director.replace("Director: ", ""))
                    movie_director_2.append("--") 
                    
                elif re.search("Directors:", director):

                    splitDirector = director.split(',')

                    if splitDirector[0]:   
                        movie_director_1.append(splitDirector[0].replace("Directors: ", ""))
                    else:
                        movie_director_1.append("--")

                    if splitDirector[1]:    
                        movie_director_2.append(splitDirector[1].replace("Directors: ", ""))
                    else:
                        movie_director_2.append("--")    

                else:
                    movie_director_1.append("--")
                    movie_director_2.append("--") 
        
        
    except:
        pass
    
    try:
        
        # getting movies All Stars name
        
        all_starts  = allUd.find_elements(By.XPATH, ".//p[3]")

        for i in all_starts:
            if i.text:
                star = i.text
#                 star = i.text.split(' | ')[1]

                if re.search(" | Stars:", star): 
                    star = star.split(' | ')[1]
                        
                    splitStars = star.split(',')

                    if splitStars[0]:   
                        movie_star_1.append(splitStars[0].replace("Stars: ", ""))
                    else:
                        movie_star_1.append("--")

                    if splitStars[1]:    
                        movie_star_2.append(splitStars[1].replace("Stars: ", ""))
                    else:
                        movie_star_2.append("--")    

                    if splitStars[2]:    
                        movie_star_3.append(splitStars[2].replace("Stars: ", ""))
                    else:
                        movie_star_3.append("--")  

                    if splitStars[3]:    
                        movie_star_4.append(splitStars[3].replace("Stars: ", ""))
                    else:
                        movie_star_4.append("--")    

                elif re.search("Stars:", star):

                    splitStar = star.split(',')
#                     print(star)

                    if splitStar[0]:   
                        movie_star_1.append(splitStar[0].replace("Stars: ", ""))
                    else:
                        movie_star_1.append("--")

                    if splitStar[1]:    
                        movie_star_2.append(splitStar[1].replace("Stars: ", ""))
                    else:
                        movie_star_2.append("--")    

                    if splitStar[2]:    
                        movie_star_3.append(splitStar[2].replace("Stars: ", ""))
                    else:
                        movie_star_3.append("--")  

                    if splitStar[3]:    
                        movie_star_4.append(splitStar[3].replace("Stars: ", ""))
                    else:
                        movie_star_4.append("--")    

                else:
                    movie_star_1.append("--")
                    movie_star_2.append("--") 
                    movie_star_3.append("--")
                    movie_star_4.append("--") 
        
    except:
        pass
    
    try:
        
        # getting durations name
        if len(allUd.find_elements(By.XPATH, ".//span[@class='runtime']")) > 0:
            durations  = allUd.find_elements(By.XPATH, ".//span[@class='runtime']")
            
            for i in durations:
                movie_durations.append(i.text)
                
        else:
            movie_durations.append("--")       
        
    except:
        pass
    
    try:
        
        # getting certificates name
        if len(allUd.find_elements(By.XPATH, ".//span[@class='certificate']")) > 0:
            certificates  = allUd.find_elements(By.XPATH, ".//span[@class='certificate']")
            
            for i in certificates:
                movie_certificates.append(i.text)
                
        else:
            movie_certificates.append("--")       
        
    except:
        pass
    
    try:
        
        # getting genres name
        if len(allUd.find_elements(By.XPATH, ".//span[@class='genre']")) > 0:
            genres  = allUd.find_elements(By.XPATH, ".//span[@class='genre']")
            
            for i in genres:
                movie_genres.append(i.text.replace(',', ' '))
                
        else:
            movie_genres.append("--")       
        
    except:
        pass
```
